## 필요한 모듈 불러오기

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## seaborn 모듈안에는 여러가지의 예제 데이터가 존재

```python
# tips 데이터 불러오기
tips = sns.load_dataset('tips')
```

```python
# 데이터 확인
tips.head()
```

## 1. relplot(scatter, line을 그려주는 함수)
## default : scatter

## (1) scatter plot(산점도)

```python
# x : x축 column
# y : y축 column
# data : 사용할 data frame
sns.relplot(x='total_bill',y='tip', data=tips)
```

```python
# hue(명목형) : 해당 변수를 기준으로 색깔 다르게 표시
sns.relplot(x='total_bill', y='tip',hue='sex', data=tips)
```

```python
# hue(실수형) : 해당 변수를 기준으로 색깔 그라데이션으로 표시
sns.relplot(x='total_bill', y='tip', hue='size', data=tips)
```

```python
# style: 해당 point의 marker를 다르게 표시
sns.relplot(x='total_bill', y='tip', hue='size', style='smoker', data=tips)
```

```python
# size: 점의 크기 조정
sns.relplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=tips)
```

```python
# alpha: 점의 투명도 조정(0~1)
sns.relplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=tips, alpha=0.5) ## 진한게 가려지기 때문에
```

```python
# col : 선택된 변수에 따라서 subplot을 그려준다
sns.relplot(x='total_bill', y='tip', hue='sex', col='smoker', data=tips)
# 흡연 안했을 때는 선형 데이터
```

## (2) line plot

```python
# x : x축 column
# y : y축 column
# data : 사용할 data frame
# hue : 집단별 구분
# style : 해당 point의 marker를 다르게 표시
# default가 scatter임

sns.relplot(x='tip', y='total_bill', hue='sex', style='sex', kind='line', data=tips)
```

## 2. catplot() : 변수가 명목형일 때 사용

## (1) Strip plot

```python
# kind : 'strip'
# order : 명목형 범주 지정 - 값이 너무 동떨어져 있을 때 outlier는 제외하기 위함
# jitter : 점이 겹쳐있을 때 살짝 떨어져서 보이게 만드는 기능 - 색이 똑같음, 분포만 몰려 있을 때
sns.catplot(x='smoker', y='tip', jitter= False, order=['No', 'Yes'], data=tips)
```

```python
sns.catplot(x='smoker', y='tip', data=tips)
```

## (2) boxplot

```python
# kind : 'box'
sns.catplot(x='smoker', y='tip', hue='sex', kind = 'box', data=tips)
```

## (3) barplot

```python
# kind : 'bar'
# dodge : 집단 막대 그래프를 나누어서 그림(default:True)
sns.catplot(x='smoker', y='tip', hue='sex', kind='bar', dodge = True, data=tips)
# 선 : confidence interval (신뢰구간), 정규분포 가정이 드물기 때문에 잘쓰지 않음
```

```python
sns.catplot(x='smoker', y='tip', hue='sex', kind='bar', ci=None, dodge=False, data=tips) # stack
# 나눌 범주형 column이 많을 때 누적 분포 사용
```

## 3. countplot() : 명목형 변수의 개수를 막대그래프로 시각화

```python
# matplotlib : 일일이 색깔 다르게 해야 함
sns.countplot(tips['sex'])
plt.ylabel('counts')
```
