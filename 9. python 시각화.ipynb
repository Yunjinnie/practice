## 여러개의 그래프를 하나로 그리기

```python
# figsize 지정
fig = plt.figure(figsize = (5,5))
# ax1
ax1 = fig.add_subplot(2,2,1) # 2행 2열 중 1st
# 123
# 456
# 789 순서

# ax2
ax2 = fig.add_subplot(2,2,2)

# ax3
ax3 = fig.add_subplot(2,2,3)

# ax4
ax4 = fig.add_subplot(2,2,4)

# histogram 그리기
ax1.hist(random.sample(range(1,100), 50), bins = 10, color='y', alpha = 0.3) # alpha 가 낮을 수록 투명해짐

# bar 그리기
ax2.bar(sr2.index, height = sr2.values, color='b', alpha = 0.5) 
# 투명도로 하면 색이 겹쳐지고 다른 점에 가려져서 보이지 않는 경우 방지

# scatter 그리기
ax3.scatter(random.sample(range(1,100), 50), random.sample(range(1,100), 50), color = 'Pink')
# 각각의 1열 값과 매칭이 됨 - y축과 x축의 관계를 파악하고자 할 때

# ax4 위치에 y=x 그래프를 그려보기
ax4.plot([x for x in range(10)] ) # x는 default : index 하나씩 증가

plt.subplots_adjust(hspace=0.5 , wspace=0.3) # height, width
```

## 3. 대규모 전쟁의  공격진영과 수비 진영의 숫자 시각화
### 대규모 전쟁 = 공격과 수비 모두 합쳐서 10,000명 이상 참가한 전쟁

```python
# 1. 대규모 전쟁만 선택
big_bat = bat[bat['attacker_size'] + bat['defender_size'] >=10000]
big_bat
```

```python
# 2. 필요한 컬럼만 선택
big_bat = big_bat[['name', 'attacker_size', 'defender_size']]
big_bat
```

```python
big_bat.set_index('name', inplace=True)
big_bat
```

```python
# big_battles 시각화
big_bat.plot(kind='barh', barmode = 'stack')
# horizontal
```

## 4. 년도별로 사망자의 숫자와 전쟁이 벌어진 횟수 시각화

```python
temp = dth.groupby('Death Year')['Name'].count()
temp
```

```python
temp2 = bat.groupby('year')['name'].count() # groupby에서는 value count 불가하므로 기준 변수 세워야
temp2
```

```python
temp3 = pd.concat([temp, temp2], axis = 1)
temp3.columns = ['Death', 'battles']
temp3.plot()
```
