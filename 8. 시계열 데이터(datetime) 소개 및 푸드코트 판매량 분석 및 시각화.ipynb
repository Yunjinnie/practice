## date_range(start, end, periods)
### start : 시작 기간
### end : 끝 기간
### periods : 기간
### periods는 start부터 시작해서 periods기간 만큼
### freq : 'D' or 'M' or 'Y' or 'W-요일(영어로 앞에 3글자)' or 'B'

```python
# period 쓸 때는 end 쓰지 않음
# pandas 함수
# freq : 'D'
pd.date_range(start = '19981005', end = '20200514', freq = 'D')
```

```python
# freq: 'M'
pd.date_range(start = '19981005', end = '20200514', freq = 'M')
```

```python
# freq: 'Y'
len(pd.date_range(start = '19981005', end = '20200514', freq = 'Y'))
```

```python
# freq: 'W-MON' 월요일 ... 특정 날짜만 추출
pd.date_range(start = '19981005', end = '20200514', freq = 'W-MON')
# 월요일에 태어났구나...
```

```python
# freq: 'B'
pd.date_range(start = '19981005', end = '20200514', freq = 'B')
# 주말 빼고
```

## 실습

### 1. 전체 기간 동안, 아침, 점심, 저녁 메뉴로 가장 인기 있었던 메뉴를 각각 착기

```python
#data.groupby(['dine_type'])['use_count'].apply(ascending = False)

temp = pd.DataFrame(data.groupby(['dine_type', 'menu'])['use_count'].sum().unstack(level = 0))
# 상위레벨
```

```python
temp.sort_values(by = 'breakfast', ascending = False)
```

```python
temp.sort_values(by = 'dinner', ascending = False)
```

### 2. 2014년 한 해 동안, 각 코너 별  월별 판매량 구해보기 

```python
data2014 = data['2014']
# 코너를 기준으로 월별로
data2014.pivot_table(index = 'corner', columns = 'month', aggfunc = 'sum', values = 'use_count')
# pivot 에서는 unstack 말고 columns로 추가
```

```python
plt.scatter(data2014['corner'], data2014['use_count'])
```
