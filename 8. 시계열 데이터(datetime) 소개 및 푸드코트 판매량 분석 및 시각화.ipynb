## date_range(start, end, periods)
### start : 시작 기간
### end : 끝 기간
### periods : 기간
### periods는 start부터 시작해서 periods기간 만큼
### freq : 'D' or 'M' or 'Y' or 'W-요일(영어로 앞에 3글자)' or 'B'

```python
# freq : 'D'
pd.date_range(start='19940321', end='20200514', freq='D')
```

```python
# freq: 'M'
pd.date_range(start='19940321', end='20200514', freq='M')
```

```python
# freq: 'Y'
len(pd.date_range(start='19940321', end='20200514', freq='Y'))
```

```python
# freq: 'W-MON'
pd.date_range(start='19940321', end='20200514', freq='W-MON')
```

```python
# freq: 'B'
pd.date_range(start='19940321', end='20200514', freq='B')
```

```python
# index 최대값
food.index.max()
```

```python
# 월요일만 뽑아오기
monday = pd.date_range(start=food.index.min(),
                       end=food.index.max(),
                       freq='W-MON')
monday
```

```python
# tolist() : list로 바꾸기
monday.tolist()
```

```python
# 매주 월요일 데이터만 불러오기
food[food.index.isin(monday)]
```

```python
# wday: 월화수목금을 숫자로 01234표시
food['wday'].value_counts()
```

```python
# 평일만 불러오기
weekday = pd.date_range(start=food.index.min(),
                        end=food.index.max(),
                        freq='B')
weekday
```

```python
# 실제 데이터 날짜 길이
food.index.unique().size
```

```python
# 평일 날짜 길이
weekday.size
```

```python
# 포함되지 않는 날짜 확인
weekday[~weekday.isin(food.index.unique())]
```

```python
# groupby를 사용하여 date별로 use_count의 합 구하기
pd.DataFrame(food.groupby('date')['use_count'].sum())
```

```python
# pivot_table을 사용하여 date별로 use_count의 합 구하기
food.pivot_table(index='date', aggfunc='sum', values='use_count')
```

## 실습

### 1. 전체 기간 동안, 아침, 점심, 저녁 메뉴로 가장 인기 있었던 메뉴를 각각 착기

```python
food.head()
```

```python
temp = pd.DataFrame(food.groupby(['dine_type', 'menu'])['use_count'].sum().unstack(level=0))
```

```python
# 아침
temp.sort_values(by='breakfast', ascending=False)

# 점심
temp.sort_values(by='lunch', ascending=False)

# 저녁
temp.sort_values(by='dinner', ascending=False)
```

### 2. 2014년 한 해 동안, 각 코너 별  월별 판매량 구해보기 

```python
food2014 = food['2014']
food2014.pivot_table(index='corner',columns='month', aggfunc='sum', values='use_count')
```

### 3. 매진을 가장 많이 기록한 상위 10개 메뉴와 매진횟수를 출력

```python
food_sold_out = food[food.is_sold_out == 'T']
food_sold_out['menu'].value_counts()
```
